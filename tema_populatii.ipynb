In fisierul populatii.txt sunt descrise populațiile de iepuri, lincși și morcovi din Canada, pe o perioadă de 20 de ani.

Plecând de la codul sursă de mai jos, calculați fără a folosi cicluri for/while:

```python
import numpy as np
data = np.loadtxt('populatii.txt')
an, iepure, linx, morcov = data.T  # salveaza coloanele ca si variabile
import matplotlib.pyplot as plt
plt.axes([0.2, 0.1, 0.5, 0.8])
plt.plot(an, iepure, an, linx, an, morcov)
plt.legend(('Iepure', 'Linx', 'Morcov'), loc=(1.05, 0.5))
plt.show()
```

Media și variația standard pentru fiecare specie.

```python
print("Iepuri")
print("Media:", iepure.mean(axis=0)) #axis=0 pe coloana
print("Variatia standard:", iepure.std(axis=0))
print("Lincsi")
print("Media:", linx.mean(axis=0))
print("Variatia standard:", linx.std(axis=0))
print("Morcovi")
print("Media:", morcov.mean(axis=0))
print("Variatia standard:", morcov.std(axis=0))
```

 Care specie are cea mai mare populație?

```python
populatii = data[:,1:] #toate elem fara prima coloana
maxim = np.argmax(populatii, axis=1) #axis=1 pe linie
specie = np.array(['Iepure', 'Linx', 'Morcov'])
print("Specia cu cea mai mare populație:")
print(an,specie[maxim])
```

 In ce ani, orice populație este mai mare de 50 000? (np.any)

```python
x = np.any(populatii > 50000, axis=1)
print("Orice populație mai mare de 50000 in anii:", an[x])
```

Primii 2 ani pentru fiecare specie în care au avut cele mai mici populații (argsort)

```python
y = np.argsort(populatii, axis=0)[:2] #elem de la inceput pana la stop-1
print("Primii 2 ani pentru fiecare specie în care au avut cele mai mici populații:")
print(an[y])
```

Realizati un grafic de comparație a populației de iepuri și lincși. Verificați corelația. (np.gradient, np.corrcoef)

```python
e = np.gradient(iepure, 1.0)
l = np.corrcoef(e, linx)
print("Comparație a populației de iepuri și lincși")
print(l)
```

```python
plt.plot(an, e, an, linx)
plt.show()
```
